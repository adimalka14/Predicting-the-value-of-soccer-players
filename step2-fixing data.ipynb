```python
df2.loc[df2['Weight'] == 0]
```

We have problem with the Weight because we have some rows that the value is 0.
We saw a proportion between height and weight 
The average length is 181 and the average weight is 75 approx
181/75=2.4

This way of treatment seems more logical to us than using the median or average .
Due to the fact that soccer players are usually in good physical shape and there is a direct relationship between height and weight.

```python
df2.loc[:,['Length','Weight']].describe()
```

```python
df2.loc[df2['Weight'] == 0,['Weight']] = df2.loc[df2['Weight'] == 0,['Length']]/2.4
df2.loc[:,['Length','Weight']].describe()
```
